```python
import matplotlib.pyplot as plt
import numpy as np

original_latency =  [82]*25
refined_latency = [655.60,3440.04,5576.59,888.13,2608.89,87.00,76.68,80.09,78.10,65.61,71.56,66.57,81.11, 67.62,87.42,70.70,86.46,1022.67,89.56,80.34,74.15,81.22,87.09,80.66,90.05,]

# calculate the difference between the original and refined latency
latency_difference = np.subtract(original_latency, refined_latency)
percentage_difference = (latency_difference / original_latency) * 100

# create bar graph with the latency data and difference highlighted
x = np.arange(len(original_latency))
width = 0.35

fig, ax = plt.subplots(figsize=(20,10), dpi=400)
rects1 = ax.bar(x - width/2, original_latency, width, label='$T_{orginal}$')
rects2 = ax.bar(x + width/2, refined_latency, width, label='$T_{refined}$')
rects3 = ax.bar(x, latency_difference, width, label='$T_{orginal}-T_{refined}$', alpha=0.5, color='r')

# add labels and title to the plot
ax.set_ylabel('Latency ($\\mu$s)')
ax.set_xlabel('Methodologies')
ax.set_title('Latency comparison between original and refined model')
ax.set_xticks(x)
ax.set_xticklabels(('M-1', 'M-2', 'M-3', 'M-4', 'M-5','M-6','M-7', 'M-8', 'M-9', 'M-10', 'M-11','M-12','M-13','M-14','M-15','M-16','M-17','M-18','M-19','M-20','M-21','M-22','M-23','M-24','M-25'))
ax.legend()

# add values and percentage difference to the bars
def autolabel(rects):
    for i, rect in enumerate(rects):
        height = rect.get_height()

        model_name = ax.get_xticklabels()[i].get_text()
        annotation_text = '{}%'.format(round(percentage_difference[i], 1))
        color = 'w' if height >= 0 else 'k'
        ax.annotate(annotation_text,
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', color=color,
                    bbox=dict(facecolor='red', edgecolor='none', pad=2.0))


# autolabel(rects1)
# autolabel(rects2)
autolabel(rects3)

plt.show()
```
